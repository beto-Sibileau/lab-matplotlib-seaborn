#### Write a loop to plot 4 scatter plots of the following features:

(INDEED THE STATEMENT SHOULD BE THE OTHER WAY, PROPERLY IS DEPENDENT VS INDEPENDET VARIABLE)

* Minutes Sedentary vs Steps
* Minutes Lightly Active vs Steps    
* Minutes Fairly Active vs Steps    
* Minutes Very Active vs Steps     

```python
# your code here
# listCols = ['Minutes Sedentary','Minutes Lightly Active','Minutes Fairly Active','Minutes Very Active']
labels = ['Sedentary','Lightly Active','Fairly Active','Very Active']
colorDot = ['royalblue','forestgreen','darkorange','indianred']

# prepare figure
fig = plt.figure(figsize=[12,8],constrained_layout=True)
fig.suptitle('Steps vs Minutes - Parametric analysis: intensity', size=16, weight='bold',y=1.05)
spec = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)
subtitle_prop = {'size':14,'style':'oblique'}
label_prop = {'size':14,'family':'fantasy','style':'oblique'}

for i in range(len(labels)):
    row = int(np.floor(i/2))
    col = i%2
    axis = fig.add_subplot(spec[row,col])
    axis.set_title(labels[i],fontdict=subtitle_prop)
    if i == 0:
        axis.set_xlim(0,1000)
        plt.xticks(fontsize=12)
    elif i == 1:
        axis.set_xlim(0,600)
        plt.xticks(fontsize=12)
    else:
        axis.set_xlim(0,150)
        plt.xticks(np.arange(0,151,25),fontsize=12)
    axis.set_ylim(0,25000)
    x = 'Minutes ' + labels[i]
    plot_var = fitbit.plot.scatter(x,'Steps',ax=axis,s=20,c=colorDot[i],alpha=0.8)
    plt.sca(plot_var)
    plt.yticks(fontsize=12)
    plt.xlabel('Minutes',fontdict=label_prop)
    plt.ylabel('Steps',fontdict=label_prop)
```

#### How does the bin size affect your plot?

Basically, as you add bins the frequency per bin is reduced while more detail is added to your histogram. However, if too many bins are added, you begin capturing oscillations in your data distribution and the histogram appears noisy. That being said, the data distribution in an histogram is better appreciated with a bins size selection that is neither too low nor too high.

#### Plot the correlation matrix using seaborn.

```python
# your code here
# Compute the correlation matrix
corr = titanic.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=np.bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
g = sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.xticks(fontsize=13,rotation=45)
plt.yticks(fontsize=13,rotation=-30)
# use matplotlib.colorbar.Colorbar object
cbar = g.collections[0].colorbar
# here set the labelsize by 20
cbar.ax.tick_params(labelsize=13)
```
